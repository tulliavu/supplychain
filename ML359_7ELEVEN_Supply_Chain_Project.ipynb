```python
#Cau 1
import matplotlib.pyplot as plt
import numpy as np
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', va='bottom')
x = np.array(['Nhanh nhất','Chậm nhất','Trung bình'])
y = np.array([data['Lead times'].min(),data['Lead times'].max(),data['Lead times'].mean()])
plt.xlabel("Thời gian giao hàng từ khi khách đặt đến khi nhận")
plt.ylabel("Lead times")
plt.title("Customer's Logistics Lead time")
plt.bar(x,y)
addlabels(x, y)
plt.ylim(0,35)
fig = plt.figure(figsize = (10, 5))
plt.show()
```
